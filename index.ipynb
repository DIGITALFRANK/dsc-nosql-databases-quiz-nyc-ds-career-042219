MapReduce is conceptual approach to querying and aggragating data that is employed by Hadoop.  MapReduce is ideal for working with distributed datasets, and greatly reduces querying times when working with Big Data.  The Map phase allows queries to run individually in parrallel on RDDs or smaller block of the data in the distributed system.
